```python
import matplotlib.pyplot as plt
import numpy as np
import numba as nb

eps = 1e-6

@nb.jit(nopython = True)
def w1(z):
    N1 = int(1/(0.5*eps))
    result = 0
    for k in range(N1,1,-1):
        result += 1/(k**2-k-z)
    return result

def w2(z):
    N2 = int((6/eps)**(1/3))
    result = 1
    for k in range(N2,1,-1):
        result += z/((k**2-k)*(k**2-k-z))
    return result

z = [i/100 for i in range(1,200)]
w1 = [w1(i) for i in z]
w2 = [w2(i) for i in z]
difference = [w2[i]-w1[i] for i in range(0,199)]

plt.plot(z,difference,'r')
plt.xlabel('z')
plt.ylabel("w2(z)-w1(z)")
plt.show()
```
